# Numpy Implementation

```python
import numpy as np

# Create a 3x3 Identity matrix
I = np.eye(3)

# Calculate 9I + 1
# Since 1 is added to every element of 9I, use np.full to create a 3x3 matrix with all elements as 1
result_matrix = 9 * I + np.full((3, 3), 1)

# Perform the dot product of I and (9I + 1)
dot_product = np.dot(I, result_matrix)

print("The dot product of I and 9I+1 is:")
print(dot_product)
```

# Pandas Implementation

```python
import pandas as pd

# Create a DataFrame from a dictionary
data = {
    'Name': ['Ramesh', 'Mahesh', 'Suresh'],
    'Age': [25, 30, 35],
    'City': ['Bangalore', 'Mumbai', 'Delhi']
}
df = pd.DataFrame(data)

# Display the first 2 rows of the data frame
print("First 2 rows of the data frame:")
print(df.head(2))

# Print the age column
print("\nAge column:")
print(df['Age'])

# Filter rows where age is greater than 26
filtered_df = df[df['Age'] > 26]
print("\nFiltered rows where age is greater than 26:")
print(filtered_df)

# Add a new column 'Country' with the value 'India' for all rows
df['Country'] = 'India'
print("\nDataFrame with added 'Country' column:")
print(df)

# Handling missing values:
data1 = {
    'Name': ['Ramesh', 'Mahesh', 'Suresh'],
    'Age': [25, None, 35],
    'City': ['Bangalore', 'Mumbai', 'Delhi']
}
df2 = pd.DataFrame(data1)

# Fill missing values in the 'Age' column with the mean age
mean_age = df2['Age'].mean()
df2['Age'].fillna(mean_age, inplace=True)
print("\nDataFrame with missing values filled with mean age:")
print(df2)
```

# Matplotlib Implementation

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 3, 4])
ypoints = np.array([2, 4, 6, 8])

# Plot these points without drawing a line
plt.figure(figsize=(10, 6))
plt.scatter(xpoints, ypoints, color='black', s=100, label='Points')
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

# Plotting with a star marker
plt.figure(figsize=(10, 6))
plt.plot(xpoints, ypoints, marker='*', linestyle='', color='blue', markersize=10)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

# Using fmt format: Add circular marker, red color, and dashed line
plt.figure(figsize=(10, 6))
plt.plot(xpoints, ypoints, 'ro--', markersize=10)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

# Add xlabel, ylabel, and title for the plot
plt.figure(figsize=(10, 6))
plt.plot(xpoints, ypoints, 'ro--', markersize=10)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Plot with labels and title')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

# Create a scatter plot for xpoints and ypoints
plt.figure(figsize=(10, 6))
plt.scatter(xpoints, ypoints, color='blue', s=100)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

# Set color to the scatter plot. Blue, Green, Red, and Yellow color for each point respectively
colors = ['blue', 'green', 'red', 'yellow']
plt.figure(figsize=(10, 6))
for i in range(len(xpoints)):
    plt.scatter(xpoints[i], ypoints[i], color=colors[i], s=100)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

# Miscellaneous Modules Implementation

```python
import random
import numpy as np

# Set the seed of random to 20
random.seed(20)

arr1 = np.array([1, 24, 31, 45, 73, 81, 94, 25])

# Using the random module, pick 4 different random numbers from arr1 and return their sum
random_numbers = random.sample(list(arr1), 4)
sum_of_random_numbers = sum(random_numbers)

print("Picked random numbers:", random_numbers)
print("Sum of the picked random numbers:", sum_of_random_numbers)
```
